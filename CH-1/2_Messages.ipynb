```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI as chatGemini

llm_gemini = chatGemini(model="gemini-3-flash-preview", temperature = 0)
```

```python
from langchain_core.prompts import ChatPromptTemplate

system_input = input("Give a tone: ")
user_input = input("Give a topic: ")

prompt_template = ChatPromptTemplate.from_messages([
    ("system"," Use tone {tone}"),
    ("user", "Write on the topic {topic}")
])

prompt = prompt_template.invoke({"tone" : system_input, "topic" : user_input})

llm_gemini.invoke(prompt)
```
