LLM API CALL WITH SDK (GEMINI)

```python
from dotenv import load_dotenv
# Load .env file
load_dotenv(override=True)
```

```python
from google import genai
import os
os.getenv("GEMINI_API_KEY")

# Create client (reads key from env)
client = genai.Client()

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="Explain how AI works in a few words"
)

print(response.text)
```

LLM API CALL WITH SDK (GEMINI) through LangChain

```python
from langchain_google_genai import ChatGoogleGenerativeAI as chatGemini

llm_gemini = chatGemini(model="gemini-3-flash-preview", temperature = 0)

llm_gemini.invoke("Explain how AI works in a few words")
```
